This loads in resources

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib notebook
```

Now load the data

```python
blankfilename = 'blank.txt'
obsfilename = '1ml1.txt'
```

```python
blankdata = np.genfromtxt(blankfilename)
obsdata = np.genfromtxt(obsfilename)
wavelen = blankdata[:,0]
blank = blankdata[:,1]
obs = obsdata[:,1]
```

```python
# This graphs the intensity data
plt.figure()
plt.plot(wavelen,blank)
plt.plot(wavelen,obs)
```

```python
# This graphs the absorbance
plt.figure()
plt.plot(wavelen,-np.log(obs/blank))
```
